MORE PRACTICE WITH SEABORN


LINE PLOTS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris") #load some load_dataset
data.head()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("sepal_length VS sepal_width")
```

```python
#using xlim() & ylim()
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.xlim(5)
plt.ylim(2.5)
plt.title("sepal_length VS sepal_width")
```

```python
#Changing Figure Aesthetic using set_style();

def plot():
  sns.lineplot(x="sepal_length", y="sepal_width", data=data)
  sns.set_style("dark") #changing the theme into dark
  sns.despine() #removes the spine

with sns.axes_style('darkgrid'): #to add grids
  plt.subplot(211)
  plot()

plt.title("sepal_length VS sepal_width")
```

```python
#Color Palettes

palette = sns.color_palette()
sns.palplot(palette)
```

```python
#Diverging color palettes
palette = sns.color_palette('PiYG', 11)
sns.palplot(palette)
```

```python
#Sequential color palette
palette = sns.color_palette('Reds', 11)
sns.palplot(palette)
```

MULTIPLE PLOTS

```python
data = sns.load_dataset("iris")

def graph():
  sns.lineplot( x="sepal_length", y="sepal_width", data=data)

fig = plt.figure(figsize =(5,4))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
graph()

ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
graph()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")

def graph():
  sns.lineplot(x="sepal_length", y="sepal_width", data=data)

plt.subplot(121)
graph()

plt.subplot(122)
graph()
```

```python
def graph():
  sns.lineplot(x="sepal_length", y="sepal_width", data=data)

axes1 = plt.subplot2grid(
    (7,1),(0,0), rowspan = 2, colspan = 1
)
graph()

axes2 = plt.subplot2grid(
    (7,1),(2,0), rowspan = 2, colspan = 1
)
graph()

axes3 = plt.subplot2grid(
    (7,1),(4,0), rowspan = 2, colspan = 1
)
graph()
```

```python
plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")
plt.show()
```

```python
data = sns.load_dataset("flights")
plot = sns.PairGrid(data)
plot.map(plt.plot)
plt.show()
```
