```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
#filling the missing values
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
# fit() method is used to train the model using the data set
reg.fit(df[['area', 'bedrooms','age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3 , 40]])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience.unique()
```

```python
import math

a = 0
for i in df.experience:
  if i == "five":
    a = 5
  elif i == 'two':
    a = 2
  elif i == 'seven':
    a = 7
  elif i == 'three':
    a = 3
  elif i == 'ten':
    a = 10
  elif i == 'eleven':
    a = 11
  else:
    a = 0
  df['experience'] = df['experience'].replace(i,a)

```

```python
df
```

```python
import math
median_testScore = df['test_score(out of 10)'].median()
median_testScore
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_testScore)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[12, 10, 10]])
```
